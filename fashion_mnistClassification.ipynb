```
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```
data=keras.datasets.fashion_mnist
```

```
(train_images, train_labels), (test_images, test_labels) = data.load_data()
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
print(train_labels[0])
```

```
print(class_names[train_labels[0]])
plt.imshow(train_images[2])
```

```
print(train_images[7])
plt.imshow(train_images[7], cmap=plt.cm.binary)
```

```
train_images=train_images/255.0
test_images=test_images/255.0
```

```
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(28,28)),
                        keras.layers.Dense(128, activation="relu"),
                        keras.layers.Dense(10, activation="softmax")
])
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.fit(train_images, train_labels, epochs=5)
```

```
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"Loss: {test_loss} Acc: {test_acc}")
```

```
prediction=model.predict(test_images)
for i in range(10):
  print(f"Actual: {class_names[test_labels[i]]} Predicted: {class_names[np.argmax(prediction[i])]}")
  plt.imshow(test_images[i])
  plt.xlabel(f"Actual: {class_names[test_labels[i]]}")
  plt.ylabel(f"Predicted: {class_names[np.argmax(prediction[i])]}")
  plt.title("SHOW")
```
